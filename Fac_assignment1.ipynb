```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

ticker2 = yf.Ticker("AAPL")
stock_data2 = yf.download("AAPL", start="2024-01-01", end="2024-11-30")

ticker3 = yf.Ticker("MSFT")
stock_data3 = yf.download("MSFT", start="2024-01-01", end="2024-11-30")

ticker4 = yf.Ticker("GOOGL")
stock_data4 = yf.download("GOOGL", start="2024-01-01", end="2024-11-30")

initial_investment = 10000

initial_price = stock_data2['Adj Close'].iloc[0]
shares_purchased = initial_investment / initial_price

portfolio_value = stock_data2['Adj Close'] * shares_purchased

daily_returns = portfolio_value.pct_change()

initial_value = portfolio_value.iloc[0]
final_value = portfolio_value.iloc[-1]
total_return = (final_value - initial_value) / initial_value * 100

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
portfolio_value.plot(title='AAPL Portfolio Value')
plt.xlabel('Date')
plt.ylabel('Portfolio Value ($)')

plt.subplot(2, 1, 2)
daily_returns.plot(title='AAPL Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')

plt.tight_layout()
plt.show()

print(f"Stock: AAPL")
print(f"Initial Investment: ${initial_investment:.2f}")
print(f"Initial Price: ${initial_price.iloc[0]:.2f}")
print(f"Final Portfolio Value: ${final_value.iloc[0]:.2f}")
print(f"Total Return: {total_return.iloc[0]:.2f}%")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

def analyze_portfolio(tickers, weights, start_date, end_date, initial_investment=10000):
    stock_data = {}

    for ticker in tickers:
        stock_data[ticker] = yf.download(ticker, start=start_date, end=end_date)

    portfolio_performance = {}

    total_portfolio_value = pd.DataFrame()

    plt.figure(figsize=(12, 6))

    for i, ticker in enumerate(tickers):
        data = stock_data[ticker]
        investment_per_stock = initial_investment * weights[i]
        initial_price = data['Adj Close'].iloc[0]
        shares_purchased = investment_per_stock / initial_price

        portfolio_value = data['Adj Close'] * shares_purchased

        total_portfolio_value[ticker] = portfolio_value

        daily_returns = portfolio_value.pct_change()
        initial_value = portfolio_value.iloc[0]
        final_value = portfolio_value.iloc[-1]
        total_return = (final_value - initial_value) / initial_value * 100

        portfolio_performance[ticker] = {
            'initial_investment': investment_per_stock,
            'initial_price': initial_price,
            'shares_purchased': shares_purchased,
            'final_value': final_value,
            'total_return': total_return,
        }

        plt.subplot(len(tickers), 2, 2*i+1)
        portfolio_value.plot(title=f'{ticker} Portfolio Value')
        plt.xlabel('Date')
        plt.ylabel('Portfolio Value')

        plt.subplot(len(tickers), 2, 2*i+2)
        daily_returns.plot(title=f'{ticker} Daily Returns')
        plt.xlabel('Date')
        plt.ylabel('Daily Return')

    total_portfolio_value['Total'] = total_portfolio_value.sum(axis=1)

    plt.show()

    print("Portfolio Performance Summary:")
    for ticker, performance in portfolio_performance.items():
        print(f"\nStock: {ticker}")
        print(f"Initial Investment: ${performance['initial_investment']:.2f}")
        print(f"Initial Price: ${performance['initial_price'].iloc[0]:.2f}")
        print(f"Final Stock Value: ${performance['final_value'].iloc[0]:.2f}")
        print(f"Total Return: {performance['total_return'].iloc[0]:.2f}%")

    initial_total_value = total_portfolio_value['Total'].iloc[0]
    final_total_value = total_portfolio_value['Total'].iloc[-1]
    total_portfolio_return = (final_total_value - initial_total_value) / initial_total_value * 100

    print("\nTotal Portfolio Summary:")
    print(f"Initial Total Portfolio Value: ${initial_total_value:.2f}")
    print(f"Final Total Portfolio Value: ${final_total_value:.2f}")
    print(f"Total Portfolio Return: {total_portfolio_return:.2f}%")

    return {
        'portfolio_performance': portfolio_performance,
        'total_portfolio_value': total_portfolio_value
    }

```

```python
#Example
tickers = ['AAPL', 'MSFT', 'GOOGL', 'TSLA', 'NVDA']
weights = [0.3, 0.3, 0.2, 0.2, 0.1]
start_date = '2024-01-01'
end_date = '2024-11-30'

result = analyze_portfolio(tickers, weights, start_date, end_date)
```
